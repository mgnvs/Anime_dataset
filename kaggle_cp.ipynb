```python
distribuicao_pontuacao = df['Score']
print("Média da pontuação:", np.mean(distribuicao_pontuacao))
print("Desvio padrão da pontuação:", np.std(distribuicao_pontuacao))
```

```python
contagem_generos = df['Genres'].value_counts()
contagem_generos
```

```python
top_10_score = df.nlargest(10, 'Score')
print(top_10_score[['Title', 'Score']])
```

```python
top_10_popularity = df.nsmallest(10, 'Popularity')
print(top_10_popularity[['Title', 'Popularity']])
```

```python
df['Rank'] = pd.to_numeric(df['Rank'], errors='coerce')
```

```python
top_10_rank = df.nsmallest(10, 'Rank')
top_10_rank[['Title', 'Rank']]
```

```python
# 1. Análise de Diferentes Tipos de Gêneros
# Contagem de ocorrências de diferentes gêneros
genre_counts = df['Genres'].str.split(', ', expand=True).stack().value_counts()
"Contagem de ocorrências de diferentes gêneros:"
genre_counts
```
